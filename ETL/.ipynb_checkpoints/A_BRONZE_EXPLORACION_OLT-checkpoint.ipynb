# CONEXION OLTS

```python
import pandas as pd #Importar librería manipulación datos
import re #Importar librería expresiones regulares
import os #Importar módulo sistema operativo
import json  #Importar módulo para lectura de JSON
import time #Importar módulo tiempo
from netmiko import ConnectHandler #Importar cliente conexión SSH
from datetime import datetime #Importar módulo fechas

# Cargar configuración desde JSON
with open('config_etl_olts.json', 'r') as file: #Abrir archivo JSON
    config = json.load(file) #Cargar datos del JSON

INVENTARIO_OLTS = config['inventario_olts'] #Asignar diccionario OLTs

#Definir ruta salida
RUTA_SALIDA = "datalake/bronze/huawei_inventory"
if not os.path.exists(RUTA_SALIDA): #Verificar existencia directorio
    os.makedirs(RUTA_SALIDA) #Crear directorio si falta

str_yymm = datetime.now().strftime('%y%m') #Extraer año y mes actual
```

# CREACION DATAFRAMES

```python
def obtener_info_base(texto_raw): #Definir función parseo info
    data_dict = {} #Inicializar diccionario datos
    lines = texto_raw.splitlines() #Dividir texto líneas
    modo_descripcion = False #Inicializar bandera modo
    
    #Compilar regex tabla principal SN
    regex_main = re.compile(r'^\s*(\d+/\s*\d+/\s*\d+)\s+(\d+)\s+([A-F0-9]{12,16})\s+\S+\s+(\w+)\s+')
    #Compilar regex tabla descripción
    regex_desc = re.compile(r'^\s*(\d+/\s*\d+/\s*\d+)\s+(\d+)\s+(.*)$')

    for line in lines: #Iterar líneas texto
        line = line.strip() #Limpiar espacios extremos
        if not line: continue #Saltar líneas vacías
        
        #Detectar cabecera descripción
        if "Description" in line and "ONT-ID" in line: #Validar cambio tabla
            modo_descripcion = True #Activar modo descripción
            continue #Saltar línea cabecera
            
        if not modo_descripcion: #Procesar tabla principal
            m = regex_main.search(line) #Buscar coincidencia regex
            if m: #Validar coincidencia encontrada
                fsp = m.group(1).replace(" ", "") #Eliminar espacios blancos
                ont_id = int(m.group(2)) #Convertir ID entero
                sn = m.group(3) #Extraer número serie
                estado = m.group(4).lower() #Extraer estado minúsculas
                
                if ont_id not in data_dict: #Verificar existencia ID
                    data_dict[ont_id] = {
                        'interface_id': fsp,
                        'ont_id': ont_id,
                        'serial_number': sn,
                        'status_olt': estado,
                        'description': '' #Inicializar descripción vacía
                    }
        else: #Procesar tabla descripción
            m = regex_desc.search(line) #Buscar coincidencia regex
            if m: #Validar coincidencia encontrada
                ont_id = int(m.group(2)) #Extraer ID entero
                desc = m.group(3).strip() #Extraer descripción limpia
                if ont_id in data_dict: #Verificar existencia ID
                    data_dict[ont_id]['description'] = desc #Asignar descripción

    return pd.DataFrame(list(data_dict.values())) #Retornar dataframe info

def obtener_info_version(texto_raw): #Definir función parseo versión
    registros = [] #Inicializar lista registros
    #Regex Ajustado: (F)/(S)/(P)/(ID) Vendor Model Version
    regex = re.compile(r'(\d+)\s*/\s*(\d+)\s*/\s*(\d+)\s*/\s*(\d+)\s+(\S+)\s+(\S+)\s+(\S+)')
    
    lines = texto_raw.splitlines() #Dividir texto líneas
    for line in lines: #Iterar líneas texto
        line = line.strip() #Limpiar espacios extremos
        if "Vendor" in line: continue #Saltar cabecera tabla
        if "----" in line: continue #Saltar separadores
        
        m = regex.search(line) #Buscar coincidencia regex
        if m: #Validar coincidencia encontrada
            fsp = f"{m.group(1)}/{m.group(2)}/{m.group(3)}" #Formatear interfaz
            ont_id = int(m.group(4)) #Extraer ID ont
            vendor = m.group(5) #Extraer vendor
            modelo = m.group(6) #Extraer modelo ont
            version = m.group(7) #Extraer versión software
            
            if vendor != '-' and modelo != '-': #Validar datos útiles
                registros.append({
                    'interface_id': fsp,
                    'ont_id': ont_id,
                    'ont_model': modelo,
                    'software_version': version
                }) #Agregar registro lista

    return pd.DataFrame(registros) #Retornar dataframe versión
```

# UNIR DATAFRAMES Y GENERAR PARQUET

```python
def pipeline_unificado(nombre, credenciales): #Definir función principal
    print(f"[{nombre}] Iniciando sesión única...") #Imprimir mensaje inicio
    
    device = {
        'device_type': 'huawei_olt',
        'host': credenciales['ip'],
        'username': credenciales['user'],
        'password': credenciales['pass'],
        'port': credenciales['port'],
        'fast_cli': False,
        'global_delay_factor': 2
    } #Configurar parámetros conexión
    
    try: #Iniciar bloque manejo errores
        conn = ConnectHandler(**device) #Establecer conexión SSH
        #conn = ConnectHandler(device_type='huawei', **credenciales)
        conn.enable() #Habilitar modo privilegiado

        #conn.send_command("scroll", expect_string=r"#") # En algunos Huawei VRP antiguos
        #conn.send_command("screen-length 0 temporary", expect_string=r"#") # En la mayoría de Huawei modernos
        
        #Configuración inicial
        print(f"[{nombre}] Configurando terminal...") #Imprimir estado
        conn.config_mode() #Entrar modo configuración
        conn.send_command("undo smart", expect_string=r"#") #Desactivar smart paging
        conn.exit_config_mode() #Salir modo configuración
        
        #Detectar tarjetas activas
        print(f"[{nombre}] Detectando tarjetas...") #Imprimir estado
        out_board = conn.send_command("display board 0") #Consultar tarjetas
        slots = [] #Inicializar lista slots
        for line in out_board.splitlines(): #Iterar líneas salida
            if "GP" in line and "Normal" in line: #Validar condiciones tarjeta
                parts = line.split() #Dividir línea columnas
                if parts[0].isdigit(): slots.append(parts[0]) #Agregar número slot
        print(f"[{nombre}] Slots detectados: {slots}") #Imprimir slots encontrados
        
        #Listas para almacenar dataframes temporales
        lista_dfs_info = [] #Inicializar lista info
        lista_dfs_version = [] #Inicializar lista versión
        
        #OBTENER INFO BASE (Entrando a Interfaz)
        print(f"[{nombre}] FASE 1: Extracción Info Base...") #Imprimir inicio fase
        for slot in slots: #Iterar slots para info
            conn.config_mode() #Asegurar modo configuración
            res = conn.send_command_timing(f"interface gpon 0/{slot}") #Entrar interfaz gpon
            
            if "Error" in res: #Validar error entrada
                conn.exit_config_mode() #Salir configuración
                continue #Saltar siguiente slot
                
            for port in range(16): #Iterar puertos 0-15
                cmd = f"display ont info {port} all" #Definir comando info
                out_info = conn.send_command(cmd, read_timeout=90) #Ejecutar comando
                if "No ONT" in out_info or "Failure" in out_info: continue #Validar respuesta
                
                df_temp = obtener_info_base(out_info) #Parsear salida
                if not df_temp.empty: lista_dfs_info.append(df_temp) #Guardar dataframe
            
            conn.send_command_timing("quit") #Salir interfaz
            conn.exit_config_mode() #Regresar config
            
        #OBTENER VERSION (Comando Global)
        print(f"[{nombre}] FASE 2: Extracción Versiones...") #Imprimir inicio fase
        for slot in slots: #Iterar slots para versión
            for port in range(16): #Iterar puertos 0-15
                #Usar comando global display ont version 0 slot port all
                cmd = f"display ont version 0 {slot} {port} all" #Definir comando versión
                out_ver = conn.send_command(cmd, read_timeout=90) #Ejecutar comando
                if "No ONT" in out_ver or "Failure" in out_ver or "Parameter" in out_ver: continue #Validar respuesta
                
                df_temp = obtener_info_version(out_ver) #Parsear salida
                if not df_temp.empty: lista_dfs_version.append(df_temp) #Guardar dataframe
        
        conn.disconnect() #Cerrar conexión SSH
        
        #UNIFICACIÓN Y GUARDADO
        print(f"[{nombre}] Procesando datos en memoria...") #Imprimir estado
        
        if not lista_dfs_info: #Validar existencia datos info
            print(f"[{nombre}] ALERTA: No se encontraron datos base.") #Imprimir alerta
            return #Finalizar
            
        #Crear DataFrame Maestro Info
        df_master_info = pd.concat(lista_dfs_info, ignore_index=True) #Concatenar infos
        print(f"[{nombre}] Registros Base: {len(df_master_info)}") #Imprimir conteo
        
        #Crear DataFrame Maestro Versión
        if lista_dfs_version: #Verificar existencia versiones
            df_master_version = pd.concat(lista_dfs_version, ignore_index=True) #Concatenar versiones
            print(f"[{nombre}] Registros Versión: {len(df_master_version)}") #Imprimir conteo
            
            #Merge Left: Base + Version
            df_final = pd.merge(
                df_master_info,
                df_master_version,
                on=['interface_id', 'ont_id'],
                how='left'
            ) #Unir dataframes
        else: #Caso sin versiones
            print(f"[{nombre}] ALERTA: No se obtuvieron versiones.") #Imprimir alerta
            df_final = df_master_info #Asignar solo info
            df_final['ont_model'] = None #Crear columna vacía
            df_final['software_version'] = None #Crear columna vacía
            
        #Seleccionar columnas finales
        cols_finales = [
            'interface_id', 'ont_id', 'serial_number', 
            'status_olt', 'description', 'ont_model', 'software_version'
        ] #Definir orden columnas
        
        #Filtrar columnas existentes para seguridad
        cols_existentes = [c for c in cols_finales if c in df_final.columns] #Filtrar columnas
        df_final = df_final[cols_existentes] #Reordenar dataframe
        
        #Guardar Parquet Único
        archivo = f"{RUTA_SALIDA}/inventory_master_{nombre}_{str_yymm}.parquet" #Definir nombre archivo
        df_final.to_parquet(archivo, index=False) #Exportar parquet
        
        print(f"[{nombre}] EXITO TOTAL. Archivo: {archivo}") #Imprimir éxito
        print(df_final.head()) #Imprimir muestra
        
    except Exception as e: #Capturar excepción general
        print(f"[{nombre}] ERROR CRÍTICO: {e}") #Imprimir mensaje error

for nombre, creds in INVENTARIO_OLTS.items(): #Iterar OLTs
    pipeline_unificado(nombre, creds) #Ejecutar pipeline
    time.sleep(5) #Esperar tiempo seguridad
```

# MOSTRAR PARQUET

```python
import glob #Importar módulo búsqueda archivos

#Definir ruta búsqueda
RUTA_BRONZE = "datalake/bronze/huawei_inventory"

#Buscar archivos maestros generados
archivos_master = glob.glob(f"{RUTA_BRONZE}/inventory_master_*.parquet")

print(f"Archivos encontrados: {len(archivos_master)}\n") #Imprimir cantidad archivos

#Inicializar lista para resumen final
resumen_global = [] #Crear lista vacía

#Iterar sobre cada archivo encontrado
for archivo in archivos_master: #Bucle archivos
    nombre_archivo = os.path.basename(archivo) #Obtener nombre base
    print("="*80) #Imprimir separador
    print(f"ARCHIVO: {nombre_archivo}") #Imprimir nombre archivo
    
    try: #Iniciar bloque lectura
        #Leer archivo parquet
        df = pd.read_parquet(archivo) #Cargar dataframe
        
        #Calcular métricas
        total_regs = len(df) #Contar total registros
        #Contar modelos vacíos (nulos o vacíos)
        if 'ont_model' in df.columns: #Validar existencia columna
            modelos_vacios = df['ont_model'].isnull().sum() #Sumar nulos
        else:
            modelos_vacios = total_regs #Asumir todos vacíos si no existe columna
            
        #Guardar datos para resumen
        resumen_global.append({
            'Archivo': nombre_archivo,
            'Total': total_regs,
            'Modelos_Vacios': modelos_vacios
        }) #Agregar diccionario a lista
        
        #Imprimir métricas básicas
        print(f"   Total Registros: {total_regs}") #Imprimir total filas
        print(f"   Modelos Vacíos: {modelos_vacios}") #Imprimir conteo vacíos
        print(f"   Columnas: {list(df.columns)}") #Imprimir lista columnas
        
        #Mostrar primeros 5 registros
        print("\n   VISTA PREVIA (TOP 5):") #Imprimir cabecera
        if not df.empty: #Validar dataframe no vacío
            #Seleccionar columnas clave para visualización limpia
            cols_view = [
                'interface_id', 'ont_id', 'serial_number', 
                'status_olt', 'description', 'ont_model', 'software_version'
            ] #Definir columnas vista
            
            #Filtrar solo columnas existentes
            cols_final = [c for c in cols_view if c in df.columns] #Filtrar columnas
            
            #Imprimir tabla sin índice
            print(df[cols_final].head(5).to_string(index=False)) #Mostrar datos
        else: #Caso archivo vacío
            print("   [ARCHIVO VACÍO]") #Imprimir alerta
            
    except Exception as e: #Capturar error lectura
        print(f"   ERROR LEYENDO ARCHIVO: {e}") #Imprimir error

#==========================================
#RESUMEN FINAL CONSOLIDADO
#==========================================
print("\n" + "="*80) #Imprimir separador
print("RESUMEN GENERAL DE EXTRACCIÓN") #Imprimir título resumen
print(f"{'ARCHIVO':<50} {'TOTAL':<10} {'SIN MODELO':<10}") #Imprimir cabecera tabla
print("-" * 80) #Imprimir línea divisoria

total_global_onts = 0 #Inicializar contador global
total_global_vacios = 0 #Inicializar contador vacíos global

for item in resumen_global: #Iterar lista resumen
    print(f"{item['Archivo']:<50} {item['Total']:<10} {item['Modelos_Vacios']:<10}") #Imprimir fila
    total_global_onts += item['Total'] #Sumar al total global
    total_global_vacios += item['Modelos_Vacios'] #Sumar al total vacíos

print("-" * 80) #Imprimir línea cierre
print(f"{'TOTAL GLOBAL':<50} {total_global_onts:<10} {total_global_vacios:<10}") #Imprimir totales finales
print("="*80) #Imprimir cierre final
print("Revisión Finalizada.") #Imprimir fin proceso
```
