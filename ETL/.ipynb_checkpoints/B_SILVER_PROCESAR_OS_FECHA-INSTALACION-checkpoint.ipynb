# UNIR REPORTES DE SISTEMA DE COBRANZA, GENERAR NUEVA COLUMNA 'DIAS DE ANTIGUEDAD' Y 'DIAS ATENCION'

```python
import pandas as pd #Importar librería pandas
import os #Importar módulo sistema
import json #Importar módulo json
from datetime import datetime #Importar formateo de fechas

# Cargar configuración desde JSON
with open('config_etl_silver.json', 'r') as file: #Abrir archivo JSON
    config = json.load(file) #Cargar datos del JSON

extr_anio = config['etl_params']['anio'] #Definir año a procesar desde JSON
extr_mes = config['etl_params']['mes'] #Definir mes a procesar desde JSON

# Construir el formato AAMM (ej. 2026, mes 2 -> 2602)
str_yymm = datetime(extr_anio, extr_mes, 1).strftime('%y%m') #Formatear fecha

RUTA_BRONZE_ORDERS = "datalake/bronze/service_orders" #Definir ruta origen
RUTA_SILVER = "datalake/silver/service_orders" #Definir ruta destino

if not os.path.exists(RUTA_SILVER): #Verificar existencia directorio
    os.makedirs(RUTA_SILVER) #Crear directorio destino

# Construir nombres de archivo de entrada con AAMM
FILE_ORDENES = f"{RUTA_BRONZE_ORDERS}/reporte_ordenes_{str_yymm}.xlsx" #Definir archivo órdenes dinámico
FILE_ABONADOS = f"{RUTA_BRONZE_ORDERS}/reporte_abonados_{str_yymm}.xlsx" #Definir archivo abonados dinámico

def generar_silver_servicios(): #Definir función principal
    print(f"Iniciando Fusión de Servicios (Silver) para el periodo {str_yymm}...") #Imprimir inicio proceso

    try: #Iniciar bloque manejo de errores
        print("   Cargando Órdenes (Excel)...") #Imprimir estado carga
        df_ord = pd.read_excel(FILE_ORDENES) #Cargar excel órdenes
        df_ord.columns = df_ord.columns.str.strip().str.upper() #Normalizar columnas mayúsculas
        
        #Definir columnas requeridas (Ordenes)
        cols_ord = [ #Iniciar lista columnas
            'CONTRATO', 'FECHA_PEDIDO', 'MOTIVO_PEDIDO', 'PAQUETE', #Listar campos
            'DETALLE_PEDIDO1', 'DETALLE_PEDIDO2', 'FECHA_CUMPLIMIENTO', #Listar campos
            'COD_TECNICO', 'ESTATUS', 'ZONA', 'MODEM(S)' #Listar campos
        ] #Cerrar lista columnas
        
        #Filtrar solo columnas deseadas
        df_ord = df_ord[cols_ord].copy() #Filtrar columnas existentes

        print("   Cargando Abonados (Excel)...") #Imprimir estado abonados
        df_abo = pd.read_excel(FILE_ABONADOS) #Cargar excel abonados
        df_abo.columns = df_abo.columns.str.strip().str.upper() #Normalizar columnas mayúsculas
        
        #Definir columnas requeridas (Abonados)
        cols_abo = ['CONTRATO', 'ESTATUS', 'PAQUETE', 'FECHA_INSTALACION', 'MODEM(S)'] #Listar columnas abonados
        df_abo = df_abo[cols_abo].copy() #Filtrar columnas existentes

        #Asegurar mismo tipo de dato para el cruce
        df_ord['CONTRATO'] = df_ord['CONTRATO'].astype(str).str.replace(r'\.0$', '', regex=True).str.strip() #Limpiar contrato órdenes
        df_abo['CONTRATO'] = df_abo['CONTRATO'].astype(str).str.replace(r'\.0$', '', regex=True).str.strip() #Limpiar contrato abonados

        print("   Uniendo fuentes...") #Imprimir estado unión
        #Pandas manejará duplicados de nombre (PAQUETE) agregando sufijos _x y _y
        df_silver = pd.merge(df_ord, df_abo, on='CONTRATO', how='left') #Ejecutar fusión dataframes

        print("   Calculando métricas de tiempo...") #Imprimir estado cálculo
        #Convertir a datetime para poder restar fechas (Maneja errores con coerce)
        df_silver['FECHA_PEDIDO'] = pd.to_datetime(df_silver['FECHA_PEDIDO'], errors='coerce') #Convertir fecha pedido
        df_silver['FECHA_INSTALACION'] = pd.to_datetime(df_silver['FECHA_INSTALACION'], errors='coerce') #Convertir fecha instalación
        df_silver['FECHA_CUMPLIMIENTO'] = pd.to_datetime(df_silver['FECHA_CUMPLIMIENTO'], errors='coerce') #Convertir fecha cumplimiento

        #Feature 1: Antigüedad (Fecha Pedido - Fecha Instalación)
        df_silver['DIAS_ANTIGUEDAD'] = (df_silver['FECHA_PEDIDO'] - df_silver['FECHA_INSTALACION']).dt.days #Calcular días antigüedad

        #Feature 2: Tiempo Atención (Fecha Cumplimiento - Fecha Pedido)
        df_silver['DIAS_ATENCION'] = (df_silver['FECHA_CUMPLIMIENTO'] - df_silver['FECHA_PEDIDO']).dt.days #Calcular días atención

        print("   Estandarizando tipos de datos...") #Imprimir estado conversión
        #Iterar columnas para convertir objetos a string y asegurar compatibilidad parquet
        for col in df_silver.columns: #Recorrer columnas
            #Si la columna es tipo objeto (mezcla números/texto), convertir a string
            if df_silver[col].dtype == 'object': #Verificar tipo objeto
                df_silver[col] = df_silver[col].astype(str) #Forzar conversión string

        # Construir nombres de archivo de salida con AAMM
        archivo_parquet = f"{RUTA_SILVER}/master_services_silver_{str_yymm}.parquet" #Definir ruta salida parquet dinámica
        df_silver.to_parquet(archivo_parquet, index=False) #Exportar archivo parquet
        
        archivo_csv = f"{RUTA_SILVER}/master_services_silver_{str_yymm}.csv" #Definir ruta csv dinámica
        df_silver.to_csv(archivo_csv, index=False, encoding='utf-8-sig') #Exportar archivo csv
        
        print(f"[EXITO] Parquet generado: {archivo_parquet}") #Imprimir mensaje éxito parquet
        print(f"[EXITO] CSV generado: {archivo_csv}") #Imprimir mensaje éxito csv
        print(f"Total Registros: {len(df_silver)}") #Imprimir total registros
        print(f"Columnas Finales: {list(df_silver.columns)}") #Imprimir lista columnas
        print("Muestra:") #Imprimir cabecera muestra
        print(df_silver.head()) #Imprimir top 5
        
    except FileNotFoundError as e: #Capturar error archivo no encontrado
        print(f"[ERROR] Archivo no encontrado: {e}") #Imprimir mensaje error
        print(f"Asegúrese de que los archivos de entrada para el periodo {str_yymm} existan en {RUTA_BRONZE_ORDERS}.") #Sugerir solución
    except Exception as e: #Capturar otros errores
        print(f"[ERROR] Ha ocurrido un error inesperado: {e}") #Imprimir mensaje error

if __name__ == "__main__": #Verificar ejecución directa
    generar_silver_servicios() #Ejecutar función principal
```
